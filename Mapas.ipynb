Trabalhando com Mapas

```python
import folium
import pandas as pd
```

```python
mapa=folium.Map([-22.9035,-43.2096],zoom_start=5,tiles='stamentoner')
```

```python
folium.Marker([-22.90000476,-43.5786345], popup ='UEZO',tooltip="Clique aqui!!",icon=folium.Icon(color='red',icon="glyphicon glyphicon-send")).add_to(mapa)
folium.Marker([-19.9166813,-43.9344931], popup ='BH',tooltip="Clique aqui!!",icon=folium.Icon(color='green')).add_to(mapa)
folium.Marker([-19.9266813,-43.934931], popup ='BH',tooltip="Clique aqui!!").add_to(mapa)
folium.Circle(radius=100,location=[-19.8166813,-43.724931],popup ='chute',color='crimson',fill=False,).add_to(mapa)

mapa.add_child(folium.LatLngPopup())
```
