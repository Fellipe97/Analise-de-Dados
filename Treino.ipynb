```python
import matplotlib.pyplot as plt

x =[]
y =[]

dataset = open('dados.txt','r')
for linha in dataset:
    linha = linha.strip()
    A, B = linha.split(';')
    x.append(int (A))
    y.append(int (B))
dataset.close()

plt.plot(x,y)
plt.show()
```
